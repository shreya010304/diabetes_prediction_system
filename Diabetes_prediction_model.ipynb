```python
import joblib
```

```python
model=joblib.load('diabetes_rf.pkl')
```

```python
import pandas as pd
newdata=pd.read_csv('newdata_diabetes.csv')
```

```python
newdata.head()
```

```python
predict=model.predict(newdata)
```

```python
predict
```

```python
newdata["Outcome"]=predict
newdata
```

```python
newdata.loc[newdata['Outcome']==1,'Outcome']='Positive'
newdata
```

```python
print("Enter the values for the diagonstic measurements")
Glucose=int(input("Enter the Glucose value:"))
BP=int(input("Enter the Blood Pressure value:"))
Insulin=int(input("Enter the Insulin value:"))
Skin_Thickness=int(input("Enter the Skin_Thickness value:"))
BMI_value=float(input("Enter the BMI value:"))
DPF=int(input("Enter the DiabetesPedigreeFunction value:"))
Age=int(input("Enter the Age value:"))
prediction=model.predict([[Glucose,BP,Skin_Thickness,Insulin,BMI_value,DPF,Age]])
if (prediction==1):
    print("THE PERSON IS DIABETIC")
else:
    print("THE PERSON IS NOT DIABETIC")
```
